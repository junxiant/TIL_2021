Conversion to submission format

```python
# ## submission wants:
# bboxx
# category_id
# image_id
# score

# type: list of dic

# [{'bbox': [100, 50, 200, 500],
#   'category_id': 1,
#   'image_id': 1,
#   'score': 0.765078},
```

```python
pic_dir = '/content/cv/challenge_5_test_dataset/images'

#cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.3 # set threshold for this model

#pred = DefaultPredictor(cfg)

list_of_dic = []

## 
for filenames in os.listdir(pic_dir):

  # get predictions for each img
  inputs = cv2.imread(pic_dir+'/'+filenames)
  outputs = pred(inputs)

  # get num of boxes for each img 
  num_boxes = len(outputs["instances"])

  # get all bbox for 1 img
  bboxes = outputs["instances"].pred_boxes.tensor.cpu().numpy().tolist()

  # loop thru all box in each img 
  for boxes in range(num_boxes):
    
    # 1 box = 1 dic
    dic = {'bbox': '',
        'category_id': '',
        'image_id': '',
        'score': ''}

    # diff dic can have same image id 
    dic['image_id'] = int(filenames.rsplit('.', 1)[0])

    ## box label 
    dic['category_id'] = outputs['instances'].pred_classes[boxes].item()

    # 1 box = 1 dic 
    dic['bbox'] = bboxes[boxes]

    # 1 box = 1 score
    dic['score'] = float(outputs["instances"].scores.cpu().numpy()[boxes])

    # add to list
    list_of_dic.append(dic)

import json

with open("/content/c5submit2.json", "w") as outf: 
    json.dump(list_of_dic, outf)
```
